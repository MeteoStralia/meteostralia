```python
df = pd.read_csv('../data/weatherAUS.csv')
df['RainTomorrow'] = df['RainTomorrow'].astype('category')
df['Date'] = pd.to_datetime(df['Date'])
df['Months'] = df['Date'].dt.month



df.dtypes
```

```python
df_temp = df[(df['Location'] == 'Sydney') | (df['Location'] == 'Perth')
             | (df['Location'] == 'AliceSprings') | (df['Location'] == 'Darwin')
             | (df['Location'] == 'Cairns') | (df['Location'] == 'NorfolkIsland')]

df_temp.sort_values(by = 'Months', inplace = True)
```

```python
px.scatter(data_frame = df_temp,
           x = 'MinTemp',
           y = 'MaxTemp',
           color = 'RainTomorrow'
           )
```

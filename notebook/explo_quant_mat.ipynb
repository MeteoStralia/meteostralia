```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('../data/weatherAUS.csv')
df['RainTomorrow'] = df['RainTomorrow'].astype('category')
df = df.select_dtypes(include = ['float64', 'category'])
df.dtypes
```

```python
df.isna().sum() / len(df)
```

```python
print(df['MinTemp'].describe())
print(df['MinTemp'].nunique())
```

```python
sns.displot(data = df, x = 'MinTemp', kde = True, bins = 80)
plt.show()

df.dropna(subset = 'MinTemp', inplace = True)
df['MinTemp'].isna().sum()
```

```python
plt.figure(figsize = (20, 5))
sns.boxplot(data = df, x = 'MinTemp')
plt.show()
```

```python
sns.scatterplot(data = df, x = 'MinTemp')
plt.show()
```
